```python
def compare(x,y):
    # if statement
    if x >y :
        print("x is greater then y" )
        # to show what is the number of x and y
        print(f"{x} is greater then {y}")
        # by adding f at the begining of the sentense (before")
        # and {} at the unknown number 
    elif x < y:
        print("x is smaller than y")
    else:
        print("x is the same as y")
```

```python
compare(5,4)
```

## Create custom images

```python
#!pip install matplotlib (if no install, cannot import)
import matplotlib.pyplot as plt
```

```python
#create a black 2D image (grayscale image)
black_img = np.zeros((200,200),dtype=np.uint8)

plt.imshow(black_img,cmap=plt.cm.gray)
plt.title("black grayscale")
plt.xticks([]),plt.yticks([]) #make x and y axis blank, no lable
plt.show()
```

```python
#create a white image

white = 255 * np.ones((200,200))
white=np.uint8(white)

plt.imshow(white,cmap=plt.cm.gray, vmin=0,vmax=256)
#for white compulsory have the vmin and vmax
plt.title("white grayscale")
plt.xticks([]),plt.yticks([]) #make x and y axis blank, no lable
plt.show()
```

```python
#create a random color image
random_color_img=np.random.randint(0,high=256,size=(200,200,3),dtype="uint8")

plt.imshow(random_color_img)
plt.title("random color")
plt.xticks([]),plt.yticks([]) #make x and y axis blank, no lable
plt.show()
```

Try to write a simple custom function to determine whether a given integer is odd or even number.


```python
def OddEven(x):
    """
    This function is to determine the odd even nature of number
    """
    if isinstance(x,int):
        if x%2==0:
            if x==0:
                statement=" is neither odd or even number."
            else:
                statement=" is even number."
        else:
                statement=" is odd number."
    else:
        statement=" is not an integer!"
            
    return print(f"{x}" + statement)
```

## Exercise 4

Create an image as shown as the following with the help of Numpy and matplotlib modules. You can arbitrarily set the dimension of the image and white circular spot at the middle.


```python
import numpy as np
import matplotlib.pyplot as plt

image_size = 200 

image = np.zeros((image_size, image_size), dtype=np.uint8)

center_x, center_y = image_size // 2, image_size // 2

radius = image_size // 4 

y, x = np.ogrid[:image_size, :image_size]
mask = (x - center_x) ** 2 + (y - center_y) ** 2 <= radius ** 2
image[mask] = 255

plt.imshow(image,cmap=plt.cm.gray, vmin=0, vmax=256)
plt.axis('off')
plt.show()
```
